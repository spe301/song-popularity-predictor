```python
import pandas as pd
df = pd.read_csv(r'C:\Users\aacjp\Downloads\Top50.csv', engine='python')
df
```

```python
T = df.drop(['Unnamed: 0','Track.Name', 'Artist.Name'], axis='columns')
T.head()
```

```python
dummies = pd.get_dummies(T.Genre)
dummies.head(5)
```

```python
merged = pd.concat([T,dummies],axis='columns')
merged.head(5)
```

```python
final = merged.drop(['Genre','pop'],axis='columns')
final.head(5)
```

```python
X = final.drop(['Popularity'],axis='columns')
X.head(5)
```

```python
y = final.Popularity
y.head(5)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
reg = LinearRegression()
reg.fit(X, y)
```
